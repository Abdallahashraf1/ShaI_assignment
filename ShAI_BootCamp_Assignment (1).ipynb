# 5. Grouped Analysis

```python
# median for Base Pay
round(df.groupby(['JobTitle'])['BasePay'].median(),3).nlargest(10)
```

```python
#median for Overtime Pay
round(df.groupby(['JobTitle'])['OvertimePay'].median(),4).nlargest(10)
```

```python
#Number of jobs in the highest-paying 30 sectors
plt.figure(figsize=(13,8))
sectors = df['JobTitle'].value_counts()[0:30]
sns.barplot(y=sectors.index, x=sectors.values, alpha=0.6)
plt.xlabel('Number of jobs', fontsize=16)
plt.ylabel("Job Title", fontsize=16)
plt.title("Number of jobs")
plt.show();
```

```python
#calculating the average TotalPayBenefits for each of the job titles. Print the top-10 earning job titles.
#making job title lowercase
df.JobTitle = df.JobTitle.str.lower()
#if job title has more than 500 data points as True False
criteria = df.JobTitle.value_counts()>500
#take job title more than 500 data points
jobtitlelist = df.JobTitle.value_counts()[criteria].reset_index()
#
df_j = df[['JobTitle', 'TotalPayBenefits']]
#
df_j = df_j[df_j.JobTitle.isin(jobtitlelist['index'])]
#
pivoted_data = df_j.pivot_table('TotalPayBenefits', index='JobTitle' , aggfunc=np.mean)
#
sorted_salaries = pivoted_data.sort_values(by='TotalPayBenefits', ascending= False)
print(sorted_salaries[:10])
```

```python
#Visualization the code (average TotalPayBenefits for top 10  job titles)
sorted_salaries[0:10].plot.barh()
```

```python
# Let’s find the employees who earn the highest totalPay amount in each year.
df.JobTitle = df.JobTitle.str.lower()
years = df.Year.unique()
print(years)
for year in years:
    sorteddata = df[df.Year==year].sort_values(by='TotalPay',ascending=False)
    person = sorteddata.EmployeeName.iloc[0]
    salary = sorteddata.TotalPay.iloc[0]
    print("In the year %d %s earned the highest salary which is %8.1f"%(year,person,salary))
```

```python
#Finding the number of unique job titles and print 10 most frequent job titles.
print("The number of unique job titles is %d"%len(np.unique(df.JobTitle)))
print("Job Title Frequency in the dataset")
print(df.JobTitle.value_counts()[0:10])
```

```python
#Visualization the code (10 most frequent job titles)
df.JobTitle.value_counts()[0:10].plot.barh()
plt.title('10 Most frequent job titles');
```

```python
#Calculating the average TotalPay values for each of the the 10 most frequent job titles
jobtitlelist = df.JobTitle.value_counts()[0:10]
data_10jobtitle = df[df.JobTitle.isin(jobtitlelist.index)]
avgsalary_10jobtitle=data_10jobtitle.groupby(by=data_10jobtitle.JobTitle).TotalPay.mean()
print(avgsalary_10jobtitle)
```

```python
#Visualization above the code ( the average TotalPay values for each of the the 10 most frequent job titles)
avgsalary_10jobtitle.plot.barh()
plt.title('Average TotalPay for the 10 Most frequent job titles');
```

# 6. Correlation Analysis

```python
# Simple Correlation Analysis
correlation = df['TotalPay'].corr(df['BasePay'])
plt.scatter(df['BasePay'], df['TotalPay'])
plt.xlabel('BasePay')
plt.ylabel('Salary')
plt.title('Correlation between Salary and BasePay')
plt.show()
print("Correlation between Salary and BasePay:", correlation)
```

```python
# Simple Correlation Analysis
correlation = df['TotalPay'].corr(df['OvertimePay'])
plt.scatter(df['OvertimePay'], df['TotalPay'])
plt.xlabel('OvertimePay')
plt.ylabel('Salary')
plt.title('Correlation between Salary and OvertimePay')
plt.show()
print("Correlation between Salary and OvertimePay:", correlation)
```

```python
# Simple Correlation Analysis
correlation = df['TotalPay'].corr(df['Benefits'])
plt.scatter(df['Benefits'], df['TotalPay'])
plt.xlabel('Benefits')
plt.ylabel('Salary')
plt.title('Correlation between Salary and Benefits')
plt.show()
print("Correlation between Salary and Benefits:", correlation)
```

```python
# creating mask
cols = df[["BasePay","OvertimePay","OtherPay","Benefits", 'TotalPay']]
mask = np.triu(np.ones_like(cols.corr()))

g=sns.heatmap(cols.corr(),annot=True ,fmt=".2f", cmap="viridis", mask=mask)
plt.xticks(rotation=25);
```
